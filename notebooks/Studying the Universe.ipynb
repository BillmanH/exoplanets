## Events

```python
c.run_query("g.V().haslabel('event').as('event').valueMap()")
event_data = pd.DataFrame(c.clean_nodes(c.res))
event_data.sort_values('time').sample(5000)
```

```python
alt.Chart(event_data.sample(5000)).mark_bar().encode(
    alt.X("time:Q", bin=alt.Bin(maxbins=30)),
    y='count()',
)
```

## Resources on Planets

```python
renewing_resources_query =f"""
   g.V().has('label','resource').has('replenish_rate').valuemap()
"""

c.run_query(renewing_resources_query)
renewing_resources = c.clean_nodes(c.res)
renewing_resources = pd.DataFrame(renewing_resources)
renewing_resources['renewal gap'] = renewing_resources['max_volume'] - renewing_resources['volume']
renewing_resources.sort_values('renewal gap')
```
