## Summary stats of planets
for purpose of prgramatically generating planets

[List of Solar System objects](https://en.wikipedia.org/wiki/List_of_Solar_System_objects_by_size#:~:text=Larger%20than%20400%20km%20%20%20%20Body,%20%202004%20%2013%20more%20rows%20)

Atmospheric data [NASA.gov](https://www.nasa.gov/sites/default/files/files/YOSS_Act_4.pdf)

```python
import pandas as pd
import numpy as np
import re, os, yaml


pd.options.mode.chained_assignment = None 
```

The dataset is limited to `Larger than 400 km`

```python
df = pd.read_excel('../../data/Solar System Data.xlsx',engine='openpyxl',skiprows=1).drop(0)
df['radius'] = df['Radius.1'].apply(float)
df['mass'] = df['Mass.1'].str.replace(',','').apply(float)
df.head()
```

filtering on planets

```python
def split_str(x):
    regex = re.compile('[^a-zA-Z]')
    y = re.sub("[^a-zA-Z]+", "", x.split('\n')[0])
    y = re.sub("xD$", "", y)
    return y

df['Body'] = df['Body'].apply(split_str)
df = df.reset_index(drop=True)
df
```

```python
def get_type(x):
    if 'gas giant' in x:
        return "gas"
    if "ice giant" in x:
        return "ice"
    if "dwarf" in x:
        return "dwarf"
    if "terrestrial" in x:
        return "terrestrial"
    if "moon" in x:
        return "moon"
    else:
        return "no type"
```

```python
df['Type.lower'] = df['Type'].apply(lambda x: str(x).lower())
df['type'] = df['Type.lower'].apply(get_type)
df['has_rings'] = df['Type.lower'].str.contains('rings')
df['is_planet'] = df['Type.lower'].str.contains('planet')
df['is_moon'] = df['Type.lower'].str.contains('moon')
pla = df.loc[df['is_planet']]
pla[['Body','Type','type']]
```

### Adding the atmosphere

```python
def get_type(x):
    y = df.loc[df['Body'].str.lower()==x.lower()]['type'].values[0]
    return y


atmo = pd.read_csv("../data/solar_system_atmosphere.csv")
atmo = atmo[atmo['Object'].str.lower().isin(pla['Body'].str.lower())]
atmo['type'] = atmo['Object'].apply(get_type)

colnames = atmo.columns
colnames = [x.lower().replace(" ","_") for x in colnames]
atmo.columns = colnames

atmo
```

```python
atmo_dist = pd.DataFrame(atmo.groupby('type').mean()).fillna(0)
atmo_dist
```

## Aggregating that for the yaml file

```python
ptypes = pd.DataFrame(pla.groupby('type').count()['Body'])
ptypes.columns = ['count']
ptypes['prob'] = ptypes['count']/14
ptypes['mass_mean'] = pd.Series(pla.groupby('type').mean()['mass'])
ptypes['mass_std'] = pd.Series(pla.groupby('type').std()['mass'])
ptypes['radius_mean'] = pd.Series(pla.groupby('type').mean()['radius'])
ptypes['radius_std'] = pd.Series(pla.groupby('type').std()['radius'])
ptypes = ptypes.join(atmo_dist)
ptypes
```

```python
def get_moon_type(x):
    if "icy" in x:
        return "ice"
    if "terrestrial" in x:
        return "terrestrial"
    else:
        return "rocky"

mon = df.loc[df['is_moon']]
mon['type'] = mon['Type'].apply(get_moon_type)
mon[['Body','Type','type','radius']]
```

Adding the planet as a column so we can look up that planets radius.

```python
def get_orbiting(x):
    words = x.split(" ")
    planames = pla['Body'].tolist()
    for word in words: 
        if word in planames:
            return word
    return "not orbiting a planet"

```

```python
pla.loc[pla['Body']=="Jupiter",'radius'].values[0]
```

```python
mon['moon_of'] = mon['Type'].apply(get_orbiting)
mon['rad_of_orbiter'] = mon['moon_of'].apply(lambda x: pla.loc[pla['Body']==x,'radius'].values[0])
mon['relative_size'] = mon['radius'] / mon['rad_of_orbiter']
mon[['Body','Type','moon_of','radius','rad_of_orbiter','relative_size']]
```

Changing here the radius to be respective of the _relative radius_, or how big the moon is relative to the planet that it orbits. This will hopefully solve the issue of having moons that are larger than the planet that it orbits. 

```python
mtypes = pd.DataFrame(mon.groupby('type').count()['Body'])
mtypes.columns = ['count']
mtypes['prob'] = mtypes['count']/len(mon)
mtypes['mass_mean'] = pd.Series(mon.groupby('type').mean()['mass'])
mtypes['mass_std'] = pd.Series(mon.groupby('type').std()['mass'])
mtypes['radius_mean'] = pd.Series(mon.groupby('type').mean()['relative_size'])
mtypes['radius_std'] = pd.Series(mon.groupby('type').std()['relative_size'])
mtypes
```

```python
moons_out = {'moon_types':mtypes.T.to_dict()}
```
